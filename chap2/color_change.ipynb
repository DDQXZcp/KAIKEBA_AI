```python
import cv2
import numpy
import matplotlib.pyplot as plt
```

# color change

```python
def my_show_BGR(img, size=(3,3)):
    plt.figure(figsize=size)
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
img_dark = cv2.imread('dark.jpg',1)
my_show_BGR(img_dark,size=(8,8))
```

# convert to YUV channel

```python
def my_show_YUV(img, size=(3,3)):
    plt.figure(figsize=size)
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_YUV2RGB))
    plt.show()
```

```python
img_YUV = cv2.cvtColor(img_dark,cv2.COLOR_BGR2YUV)
print(img_YUV.shape)
my_show_YUV(img_YUV)
```

```python
def brighter(img,y_increase,u_increase,v_increase):
    Y,U,V = cv2.split(img)
    
    Y_lim = 255-y_increase
    Y[Y>Y_lim] = 255
    Y[Y<=Y_lim] = (y_increase + Y[Y<=Y_lim]).astype(img.dtype)
    
    U_lim = 255-u_increase
    U[U>U_lim] = 255
    U[U<=U_lim] = (u_increase + U[U<=U_lim]).astype(img.dtype)
    
    V_lim = 255-v_increase
    V[V>V_lim] = 255
    V[V<=V_lim] = (v_increase + V[V<=V_lim]).astype(img.dtype)
    
    
    return cv2.merge((Y,U,V))
```

```python
img_YUV_a = brighter(img_YUV,255,0,0)
img_YUV_b = brighter(img_YUV,0,255,0)
img_YUV_c = brighter(img_YUV,0,0,255)
my_show_YUV(img_YUV_a)
my_show_YUV(img_YUV_b)
my_show_YUV(img_YUV_c)
```

# convert to HSV channel;;  crazy happy ctrl+c, ctrl + v

```python
def my_show_HSV(img, size=(3,3)):
    plt.figure(figsize=size)
    plt.imshow(cv2.cvtColor(img,cv2.COLOR_HSV2RGB))
    plt.show()
```

```python
img_HSV = cv2.cvtColor(img_dark,cv2.COLOR_BGR2HSV)
print(img_HSV.shape)
my_show_HSV(img_HSV)
```

```python
def brighter_HSV(img,h_increase,s_increase,v_increase):
    H,S,V = cv2.split(img)
    
    H_lim = 255-h_increase
    H[H>H_lim] = 255
    H[H<=H_lim] = (h_increase + H[H<=H_lim]).astype(img.dtype)
    
    S_lim = 255-s_increase
    S[S>S_lim] = 255
    S[S<=S_lim] = (s_increase + S[S<=S_lim]).astype(img.dtype)
    
    V_lim = 255-v_increase
    V[V>V_lim] = 255
    V[V<=V_lim] = (v_increase + V[V<=V_lim]).astype(img.dtype)
    
    
    return cv2.merge((H,S,V))
```

```python
img_HSV_a = brighter_HSV(img_HSV,255,0,0)
img_HSV_b = brighter_HSV(img_HSV,0,255,0)
img_HSV_c = brighter_HSV(img_HSV,0,0,255)
my_show_HSV(img_HSV_a)
my_show_HSV(img_HSV_b)
my_show_HSV(img_HSV_c)
```
